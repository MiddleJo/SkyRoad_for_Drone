```python
import numpy as np
from PathPlanning import RRT, RRTStar, Map
import matplotlib.pyplot as plt

np.random.seed(1)
```

```python
#2D example
#list of obstacles
obstacles = [[-5,25,20,35],
             [30,25,55,35],
             [45,35,55,60],
             [45,70,55,80],
             [-5,65,30,70],
             [70,40,80,70],]

#initialise environment map
bounds = np.array([0,100])
mapobs = Map(obstacles,bounds,dim = 2)
```

```python
#initialise RRT
rrt = RRT(start=np.array([80,20]),
          goal= np.array([20,50]),
          Map = mapobs, max_iter=300)
```

```python
#plan path
path = rrt.plan()

#plot the map, path 
def draw_map(rrt,path):
    fig = plt.figure()
    ax = plt.axes()
    ax.axis([bounds[0]-0.5, bounds[1]+0.5, bounds[0]-0.5, bounds[1]+0.5])

    rrt.draw_graph(ax)
    rrt.draw_path(ax,path)
    rrt.map.plotobs(ax)
    plt.show()
    
draw_map(rrt,path)
```

```python
#initialise Informed RRT*
rrt = RRTStar(start=np.array([80,20]),
              goal= np.array([20,50]),
              Map = mapobs, max_iter=300,
              goal_sample_rate = 0.1)
```

```python
#plan path
path,cost = rrt.plan()

#plot path 
draw_map(rrt,path)
```
